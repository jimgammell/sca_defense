```python
from matplotlib import pyplot as plt
from datasets.classified_mnist import ColoredMNIST, WatermarkedMNIST
import os

kwargs = {'train': True, 'root': os.path.join('.', 'downloads', 'MNIST'), 'download': True}
watermarked_mnist = WatermarkedMNIST(**kwargs)
colored_mnist = ColoredMNIST(**kwargs)
```

```python
samples_to_show = 64
dataset = watermarked_mnist
fig, axes = plt.subplots(int(samples_to_show**0.5), int(samples_to_show**0.5), figsize=(32, 32))
for (sample, _, _), ax in zip(dataset, axes.flatten()):
    ax.imshow((0.5*sample+0.5).numpy().transpose(1, 2, 0))
plt.show()
```
