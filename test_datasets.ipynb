```python
from datasets import ascad_datasets

dataset = ascad_datasets.AscadDataset()
```

```python
import numpy as np

num_examples = len(dataset)
trace_shape, label_shape = None, None
for trace, label in dataset:
    if trace_shape == None:
        trace_shape = trace.shape
    else:
        assert trace_shape == trace.shape
    if label_shape == None:
        label_shape = label.shape
    else:
        assert label_shape == label.shape
print(np.min(dataset.labels), np.max(dataset.labels))
```

```python
from matplotlib import pyplot as plt

num_traces = 5
fig, axes = plt.subplots(num_traces, 1, sharex=True, sharey=True, figsize=(4, 4*num_traces))
for ax in axes.flatten():
    example_idx = np.random.randint(len(dataset))
    trace, label = dataset[example_idx]
    ax.plot(trace, '.')
    ax.set_title('Label: %s'%(hex(label)))
```
