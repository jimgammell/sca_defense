```python
import os
from tqdm import tqdm
import numpy as np
from matplotlib import pyplot as plt
import torch
from torch import nn, optim
from models.lenet import LeNet5Classifier, LeNet5Sanitizer
from adversarial_train import *
from datasets.classified_mnist import WatermarkedMNIST
```

```python
batch_size = 512
device = 'cuda' if torch.cuda.is_available() else 'cpu'
disc_pretrain_epochs = 20
gen_pretrain_epochs = 20
train_epochs = 1000
```

```python
mnist_loc = os.path.join('.', 'downloads', 'MNIST')
train_dataset = WatermarkedMNIST(train=True, root=mnist_loc, download=True)
test_dataset = WatermarkedMNIST(train=False, root=mnist_loc, download=True)
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=8)
test_dataloader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=8)
```

```python
disc = LeNet5Classifier(output_classes=2, use_sn=True).to(device)
gen = LeNet5Sanitizer(use_sn=False).to(device)
disc_opt = optim.Adam(disc.parameters(), betas=(0.5, 0.999))
gen_opt = optim.Adam(gen.parameters())
disc_loss_fn = nn.CrossEntropyLoss()
gen_loss_fn = nn.MSELoss()
```

```python
results = {'epoch': []}
current_epoch = 1
def update_results(pretrain_disc_phase=False, pretrain_gen_phase=False):
    assert int(pretrain_disc_phase) + int(pretrain_gen_phase) <= 1
    global results, current_epoch
    print('Starting epoch %d.'%(current_epoch))
    progress_bar = tqdm(total=len(train_dataloader)+len(test_dataloader))
    if current_epoch == 0:
        rv = eval_epoch(train_dataloader, disc, gen, disc_loss_fn, gen_loss_fn, device, pretrain_disc_phase=False, pretrain_gen_phase=False, progress_bar=progress_bar)
    else:
        rv = train_epoch(train_dataloader, disc, gen, disc_opt, gen_opt, disc_loss_fn, gen_loss_fn, device, pretrain_disc_phase=pretrain_disc_phase, pretrain_gen_phase=pretrain_gen_phase, progress_bar=progress_bar)
    for key, item in rv.items():
        key = 'tr_'+key
        if not key in results.keys():
            results[key] = []
        results[key].append(item)
    rv = eval_epoch(test_dataloader, disc, gen, disc_loss_fn, gen_loss_fn, device, return_example=not (pretrain_disc_phase or pretrain_gen_phase), pretrain_disc_phase=pretrain_disc_phase, pretrain_gen_phase=pretrain_gen_phase, progress_bar=progress_bar)
    for key, item in rv.items():
        key = 'te_'+key
        if not key in results.keys():
            results[key] = []
        results[key].append(item)
    results['epoch'].append(current_epoch)
    print('Done.')
    print('Results:')
    for key, item in results.items():
        if key not in ['epoch', 'te_confusing_example', 'te_generated_example']:
            print('\t{}: {}\n'.format(key, item[-1]))
    if not (pretrain_disc_phase or pretrain_gen_phase):
        print('Confusing examples:')
        fig, axes = plt.subplots(4, 5, figsize=(20, 16))
        for eg, ax in zip(results['te_confusing_example'][-1].squeeze(), axes.flatten()):
            ax.imshow(eg, cmap='binary')
        plt.tight_layout()
        plt.show()
        print('Generated examples:')
        fig, axes = plt.subplots(4, 5, figsize=(20, 16))
        for eg, ax in zip(results['te_generated_example'][-1].squeeze(), axes.flatten()):
            ax.imshow(eg, cmap='binary')
        plt.tight_layout()
        plt.show()
    print('\n\n\n')
    current_epoch += 1

while current_epoch < disc_pretrain_epochs:
    update_results(pretrain_disc_phase=True)
while current_epoch < disc_pretrain_epochs+gen_pretrain_epochs:
    update_results(pretrain_gen_phase=True)
while current_epoch < disc_pretrain_epochs+gen_pretrain_epochs+train_epochs:
    update_results()
```
